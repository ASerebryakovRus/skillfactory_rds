# Предобработка датасета

```python
answer_ls = [] # создадим список с ответами. сюда будем добавлять ответы по мере прохождения теста



# сюда можем вписать создание новых колонок в датасете
data['income'] = data['revenue'] - data['budget']
data['release_month'] = data['release_date'].apply(lambda d: str(d).split('/')[0])
data['title_len'] = data['original_title'].apply(lambda x: len(x))
data['title_words'] = data['original_title'].apply(lambda x: len(x.split(' ')))



## Separate out cast into a separate DataFrame

empty_list = []

# Список списков разбитый по разделителю + imdb_id как посл. элемент
for i, d in data.iterrows():
    empty_list.append(d['cast'].split('|'))
    empty_list[i].append(d['imdb_id'])

# Создаём "плоский список из всего перечня значений в формате Series"
df_actors = pd.DataFrame({'imdb_id': [l[-1] for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)],
                          'actor': [e for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)]
                          })

display(df_actors.head(20))



## Separate out genres into a separate DataFrame

empty_list = []

# Список списков разбитый по разделителю + imdb_id как посл. элемент
for i, d in data.iterrows():
    empty_list.append(d['genres'].split('|'))
    empty_list[i].append(d['imdb_id'])


# Создаём "плоский список из всего перечня значений в формате Series"
df_genres = pd.DataFrame({'imdb_id': [l[-1] for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)],
                          'genre': [e for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)]
                          })


display(df_genres.head(20))



## Separate out prod. companies into a separate DataFrame

empty_list = []

# Список списков разбитый по разделителю + imdb_id как посл. элемент
for i, d in data.iterrows():
    empty_list.append(d['production_companies'].split('|'))
    empty_list[i].append(d['imdb_id'])


# Создаём "плоский список из всего перечня значений в формате Series"
df_companies = pd.DataFrame({'imdb_id': [l[-1] for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)],
                          'company': [e for l in empty_list for ix, e in enumerate(l) if ix <= (len(l) - 2)]
                          })


display(df_companies.head(20))
```
